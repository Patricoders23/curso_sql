### **INNER JOIN**



```sql
SELECT a.User_ID, a.User_Name, b.Order_ID
FROM User a 
INNER JOIN Orders b 
ON a.User_ID = b.User_ID 
```

### **LEFT JOIN**



```sql
SELECT a.User_ID, a.User_Name, b.Order_ID
FROM User a 
LEFT JOIN Orders b 
ON a.User_ID = b.User_ID 
```

### **RIGHT JOIN**



```sql
SELECT a.User_ID, a.User_Name, b.Order_ID
FROM User a 
RIGHT JOIN Orders b 
ON a.User_ID = b.User_ID 
```
